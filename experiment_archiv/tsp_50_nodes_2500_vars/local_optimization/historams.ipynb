```python
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

```python
data = [
    [-8014.0, -8032.0, -8299.0, -8043.0, -7943.0],
    [-8266.0, -8191.0, -8371.0, -8197.0, -8231.0],
]




labels = ['ordered 0.5%', "unordered 0.5%"]

```

```python
n_bins = 10

for i in range(len(data)):
    print(f'---------- Data Set {i+1} ----------')
    print(f'{labels[i]}')
    print(f'Mean: {statistics.mean(data[i])}')
    print(f'Median: {statistics.median(data[i])}')
    print(f'Variance: {statistics.variance(data[i])}')
    print(f'Standard Derv: {statistics.stdev(data[i])}')





plt.hist(data, n_bins, histtype='bar', density=False)
plt.legend(labels)
plt.ylabel('Frequency')
plt.xlabel('Fitness')
plt.show()
```
