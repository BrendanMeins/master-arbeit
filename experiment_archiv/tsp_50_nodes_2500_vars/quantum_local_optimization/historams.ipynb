```python
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

```python
data = [
    [-8307.0, -8491.0, -8203.0, -8546.0, -8433.0],
    [-9124.0, -9252.0, -9124.0, -9183.0, -9172.0],
    [-9221.0, -9224.0, -9307.0, -9260.0, -9302.0],
    [-8808.0, -8729.0, -8835.0, -9050.0, -8948.0],
    [-8749.0, -8591.0, -8807.0, -8772.0, -8733.0]

]




labels = ['random 0.5', 'segments ordered 0.5%', "segments unordered 0.5%", 'squares ordered 0.5%', "squares unordered 0.5%"]

```

```python
n_bins = 10

for i in range(len(data)):
    print(f'---------- Data Set {i+1} ----------')
    print(f'{labels[i]}')
    print(f'Mean: {statistics.mean(data[i])}')
    print(f'Median: {statistics.median(data[i])}')
    print(f'Variance: {statistics.variance(data[i])}')
    print(f'Standard Derv: {statistics.stdev(data[i])}')





plt.hist(data, n_bins, histtype='bar', density=False)
plt.legend(labels)
plt.ylabel('Frequency')
plt.xlabel('Fitness')

plt.show()
```
