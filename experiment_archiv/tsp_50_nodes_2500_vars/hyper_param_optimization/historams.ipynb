```python
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

```python
data = [
[-9221.0, -9224.0, -9307.0, -9260.0, -9302.0],
[-9175.0, -9186.0, -9096.0, -9270.0, -9107.0],
[-9185.0, -9272.0, -9125.0, -9042.0, -9192.0],
[-5091.0, -3297.0, -3519.0, -4193.0, -4817.0]
]




labels = ['p=100 gen=1000','p=10 gen=10000', 'p=1000 gen=100','p=10000 gen=10' ]

```

```python
n_bins = 30

for i in range(len(data)):
    print(f'---------- Data Set {i+1} ----------')
    print(f'{labels[i]}')
    print(f'Mean: {statistics.mean(data[i])}')
    print(f'Median: {statistics.median(data[i])}')
    print(f'Variance: {statistics.variance(data[i])}')
    print(f'Standard Derv: {statistics.stdev(data[i])}')





plt.hist(data, n_bins, histtype='bar', density=False)
plt.legend(labels)
plt.ylabel('Frequency')
plt.xlabel('Fitness')
plt.show()
```
