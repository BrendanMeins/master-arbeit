```python
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

```python
data = [
    [-8794.0, -8945.0, -8958.0, -8984.0, -8915.0],
    [-9515.0, -9474.0, -9503.0, -9484.0, -9519.0],
    [-9221.0, -9224.0, -9307.0, -9260.0, -9302.0],

]




labels = ['simulated_annealing', 'tabu search', 'qhea']

```

```python
n_bins = 10

for i in range(len(data)):
    print(f'---------- Data Set {i+1} ----------')
    print(f'{labels[i]}')
    print(f'Mean: {statistics.mean(data[i])}')
    print(f'Median: {statistics.median(data[i])}')
    print(f'Variance: {statistics.variance(data[i])}')
    print(f'Standard Derv: {statistics.stdev(data[i])}')





plt.hist(data, n_bins, histtype='bar', density=False)
plt.legend(labels)
plt.show()
```
