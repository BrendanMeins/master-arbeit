```python
import math
import numpy as np
np.random.seed(0)
import sys
import os
import scipy.sparse
np.set_printoptions(threshold=sys.maxsize)




current = os.path.dirname(os.path.realpath('__file__'))
parent = os.path.dirname(current)
parent = os.path.dirname(parent)
parent = os.path.dirname(parent)

sys.path.append(parent)

import benchmark
import tsp_qubo
import plotter
import annealing
import util
import qhea.problem
```

```python
tsp = benchmark.distances.get_tsp(benchmark.distances.get_tsp_names()[4]).distances
n_nodes = len(tsp)
penalty = 100
constraint_matrix = np.zeros((n_nodes ** 2, n_nodes ** 2))
constraint_matrix = tsp_qubo.generate.apply_penalties(constraint_matrix, penalty)

amp = np.zeros((n_nodes, n_nodes))
for i in range(len(amp)):
    amp[i][(i + 1) % n_nodes] = 1
objective = np.kron(amp, tsp)
qubo = objective + constraint_matrix
```

```python

qubo = qhea.problem.QUBO(qubo,100)
histogramm = []
for i in range(5):
    print(i)
    x = np.random.randint(0,2,qubo.n_vars)
    for j in range(500):
        print(f'iteration {j}')
        order = np.random.permutation(qubo.n_sub_problems)
        for index in order:
            subproblem = qubo.get_sub_problem(index)
            sub_qubo, d_vars = subproblem.get_qubo(x)
            solution, obj = annealing.d_wave.anneal(sub_qubo)
            x[d_vars] = solution
        print(qubo.objective_function(x))
    
    objective_value = qubo.objective_function(x)
    histogramm.append(objective_value)

print(repr(histogramm))



    



```
