```python
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

```python
data = [
    [-9221.0, -9224.0, -9307.0, -9260.0, -9302.0],
    [-9085.0, -9282.0, -9014.0, -8912.0, -9160.0]
]




labels = ['qhea', 'quantum local']

```

```python
n_bins = 20

for i in range(len(data)):
    print(f'---------- Data Set {i+1} ----------')
    print(f'{labels[i]}')
    print(f'Mean: {statistics.mean(data[i])}')
    print(f'Median: {statistics.median(data[i])}')
    print(f'Variance: {statistics.variance(data[i])}')
    print(f'Standard Derv: {statistics.stdev(data[i])}')





plt.hist(data, n_bins, histtype='bar', density=False)
plt.legend(labels)
plt.ylabel('Frequency')
plt.xlabel('Fitness')
plt.show()
```
